```python
import vbi
import numpy as np
import matplotlib.pyplot as plt
from vbi import report_cfg
from vbi import extract_features_df, extract_features_list
from vbi import get_features_by_domain, get_features_by_given_names
```

```python
D = vbi.LoadSample(nn=88)
ts = D.get_bold()
print(ts.shape)

plt.figure(figsize=(12,2.5))
plt.plot(np.arange(ts.shape[1])*0.5, ts.T, lw=1, alpha=0.5);
plt.margins(0, 0.1)
plt.xlabel("Time (s)")
plt.ylabel("BOLD signal");
```

```python
cfg = get_features_by_domain(domain="connectivity")
cfg = get_features_by_given_names(cfg, ['fc_stat'])
cfg = vbi.update_cfg(cfg, 'fc_stat', {"features":["sum"], "fc_function":"corrcoef"})
report_cfg(cfg)
```

```python
data = extract_features_list([ts], 0.5, cfg)
data
```

```python
data = extract_features_df([ts, ts], 0.5, cfg, n_workers=2)
data
```

```python
# Two module are not compatible, don't use interchangeably
try:
    from vbi.feature_extraction.features_utils_jax import get_fcd as get_fcd_jax

    fcd_jax = get_fcd_jax(ts, 1.0, win_len=10, overlap=None)
    fig, axes = plt.subplots(1,2, figsize=(7,3.5))
    axes[0].imshow(fcd_jax, aspect='auto', cmap='RdBu_r', vmin=-1, vmax=1)
    axes[0].set_title('FCD JAX')
    axes[0].set_xlabel('Time Window')
    axes[0].set_ylabel('Time Window')
    
    from vbi.feature_extraction.features_utils import get_fcd as get_fcd_numpy
    fcd_numpy = get_fcd_numpy(ts, 1.0, win_len=10, overlap=None)['full']
    axes[1].imshow(fcd_numpy, aspect='auto', cmap='RdBu_r', vmin=-1, vmax=1)
    axes[1].set_title('FCD NumPy')
    axes[1].set_xlabel('Time Window')
    axes[1].set_ylabel('Time Window')
    plt.tight_layout()
    plt.show()

    print(np.allclose(fcd_numpy, fcd_jax))

except ImportError:
    # jax is not installed
    pass
```

```python
# vmap for get_fcd
try:
    import jax
    import jax.numpy as jnp
    from functools import partial
    from vbi.feature_extraction.features_utils_jax import get_fcd as get_fcd_jax
    
    # Use partial to bind keyword arguments, then vmap over time series
    get_fcd_partial = partial(get_fcd_jax, tr=1.0, win_len=10, overlap=None)
    fcd_vmap = jax.vmap(get_fcd_partial)
    fcds = fcd_vmap(jnp.array([ts, ts, ts]))
    print(f"Vmapped FCD shape: {fcds.shape}")
except ImportError:
    # jax is not installed
    pass
```
