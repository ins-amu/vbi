## Network oscillator models

Stuart-Landau / Hopf normal-form models.  All three share the same
mathematical structure but differ in coupling type and parameter conventions.

| Model | Coupling | Parameters |
|-------|----------|------------|
| `sup_hopf` | Linear (`c_x`, `c_y`) | scalar `a`, `omega` |
| `sl` | Laplacian (`c_x - G·row_sum·x`) | scalar `a`, `omega` |
| `ghb` | Laplacian (`c_x - G·row_sum·x`) | per-node `eta`, `omega` |

### SupHopf (linear coupling)

```python
sup_hopf.describe()
```

### StuartLandau — scalar parameters, Laplacian coupling

```python
sl.describe()
```

### GHB — per-node parameters, Laplacian coupling

```python
ghb.describe()
```
